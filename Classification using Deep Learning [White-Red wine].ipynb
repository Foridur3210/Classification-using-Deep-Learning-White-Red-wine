```python
# Import pandas Library

import pandas as pd

#--- READ DATA ---# 

# Read CSV file of White Wine Quality
dataWhite = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", delimiter=';')

# Read CSV file of Red Wine Quality
dataRed = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", delimiter=';')

#--- DATA EXPLORATION ---#

# Get summary of a DataFrame Red Wine
dataWhite.info()

# Get summary of a DataFrame White Wine
dataRed.info()

# Get summary of NULL records from DataFrame Red Wine
pd.isnull(dataWhite).sum()

# Get summary of NULL records from DataFrame White Wine
pd.isnull(dataRed).sum()

# Get descriptive statistics that summarize the central tendency, dispersion and shape of a dataset’s distribution, excluding NaN values
dataRed.describe()

# Add `Winetype` column to `red` with value 1
dataRed['wine_type'] = 1

# Add `type` column to `white` with value 0
dataWhite['wine_type'] = 0

# Append `white` to `red`
data = dataRed.append(dataWhite, ignore_index=True)

# Get summary of a DataFrame
data.info()

# Create DataFrame and allocate data from old DataFrame
df=data

# Rename column names
df = df.rename(columns={'fixed acidity': 'fixed_acidity', 'volatile acidity': 'volatile_acidity', 'citric acid':'citric_acid', 'residual sugar':'residual_sugar', 'free sulfur dioxide':'free_sulfur_dioxide', 'total sulfur dioxide':'total_sulfur_dioxide'})

# Get summary of a DataFrame
df.info()

#--- DATA ANALYSIS ---#

# Get descriptive statistics
df.describe()

# Get correlation of a DataFrame
df.corr()
```

```python
#--- PLOT HISTOGRAM ---#

# import seaborn library
import seaborn as sns

# Set histogram
sns.set()
df.hist(figsize=(10,10), color='red')

# import matplotlib library
import matplotlib.pyplot as plt

# Show histogram
plt.show()
```

```python
#--- PLOT HEATMAP ---#

# import seaborn library
import seaborn as sns

# Create Heatmap of correlation using seaborn library
sns.heatmap(df.corr())
```

```python

#--- TRAINING AND TEST DATA ---#

# Select columns as independent variables and values
X = df.iloc[:,0:12]

# Check head data
X.head()

# Select column as dependent variables and values
Y = df.iloc[:,12:13]

# Check head data
Y.head()

# import library for splitting data into Training set and Testing set
from sklearn.model_selection import train_test_split

# Split data into Training set and Testing set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

X_train.shape

X_test.shape

df.shape

Y_train.shape

```

```python
#--- BUILD MODEL ---#

# Import `Sequential` from `keras.models`
from keras.models import Sequential

# Import `Dense` from `keras.layers`
from keras.layers import Dense

# Initialize the constructor
model = Sequential()

# Add an input layer 
model.add(Dense(13, activation='relu', input_shape=(12,)))

# Add one hidden layer 
model.add(Dense(8, activation='relu'))

# Add an output layer 
model.add(Dense(1, activation='sigmoid'))

# Model output shape
model.output_shape

# Model summary
model.summary()

# Model config
model.get_config()

# List all weight tensors 
model.get_weights()

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
                   
model.fit(X_train, Y_train,epochs=20, batch_size=1, verbose=1)
```

```python

#--- PREDICTION ---#

y_pred = model.predict(X_test)

yp = y_pred[:5]
yp
yp.round(2)

Y_test[:5]
```
